####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(10, 8))
plt.subplots_adjust(hspace=10)
fig, axes = plt.subplots(2, 2)

a = sns.countplot(ax= axes[0, 0],x='SEXO', data=sinasc, palette='rocket')
b = sns.countplot(ax= axes[0, 1],x='CONSULTAS', data=sinasc, palette='rocket')
c = sns.countplot(ax= axes[1, 0],x='ESTCIVMAE', data=sinasc, palette='rocket')
d = sns.countplot(ax= axes[1, 1],x='GESTACAO', data=sinasc, palette='rocket')
d.tick_params(axis="x", rotation = 30)
```
