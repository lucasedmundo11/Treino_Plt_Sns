Vamos praticar agora análise descritiva de variáveis contínuas na base do DataSUS.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
# %matplotlib notebook
```

```python
sinasc = pd.read_csv('SINASC.csv')
sinasc
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
# Seaborn
sns.boxplot(y='IDADEMAE', data=sinasc, fliersize=8, linewidth=2.0, palette='Set3')
```

```python
# Pandas
sinasc[['IDADEMAE']].boxplot(grid=False, fontsize=12)
```

####  2) Faça um boxplot do peso do bebê

```python
# Seaborn
sns.boxplot(y='PESO', data=sinasc, fliersize=7, linewidth=1.7, palette='Set1')
```

```python
# Pandas
sinasc[['PESO']].boxplot(fontsize=12)
```

####  3) Faça um histograma da idade da mãe

```python
# Seaborn
sns.displot(data = sinasc['IDADEMAE'], alpha = .8, bins = 40, element = 'step', kde = True)
```

```python
# Pandas
sinasc['IDADEMAE'].plot.hist(bins = 40, alpha=0.8)
```

#### 4) Faça um histograma do peso do bebê

```python
# Seaborn
sns.displot(data = sinasc['PESO'], alpha = .9, bins = 50, element = 'step', kde = True)
```

```python
# Pandas
sinasc['PESO'].plot.hist(bins = 50, alpha=.9)
```
